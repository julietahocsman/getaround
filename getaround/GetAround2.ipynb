```python
data = pd.read_excel('dataset.xlsx')
data = data.drop(columns = ['Unnamed: 8', 'Unnamed: 9'])
data.head(10)
```

```python
df = pd.DataFrame(columns = ['rental_id', 'car_id', 'instant_booking', 'checkin_type', 'state', 'delay_at_checkout_in_minutes', 'previous_ended_rental_id', 'delay_with_previous_rental_in_minutes', 'time_between_checkout_previous_and_checkin_next'])

for value in data.previous_ended_rental_id[:(len(data))]:
    if np.isnan(value) == False:
        df = df.append(data.loc[data.rental_id == value]).append({'time_between_checkout_previous_and_checkin_next': data.delay_with_previous_rental_in_minutes[data.previous_ended_rental_id.tolist().index(value)]}, ignore_index=True)
        
df.sort_index(inplace=True)

df
```

```python
for i in range(len(df.time_between_checkout_previous_and_checkin_next)):
    value = df.time_between_checkout_previous_and_checkin_next[i]
    if np.isnan(value) == True:
        df.time_between_checkout_previous_and_checkin_next[i] = df.time_between_checkout_previous_and_checkin_next[i+1]

df.drop(range(1, len(df), 2), inplace=True)
```

### df: All rentals that have a rental afterwards (in less than 12 hours). It can be problematic if the delay at checkout overlaps the next checkin

#### time_between_checkout_previous_and_checkin_next : the time between the checkout of the current rental and the checkin of the next one.

```python
df.set_index([pd.Index(range(len(df)))])
```

```python
df['delay_at_checkout_in_minutes'] = df['delay_at_checkout_in_minutes'].fillna(0)
```

```python
df.to_csv('rentals_with_next.csv')
```
